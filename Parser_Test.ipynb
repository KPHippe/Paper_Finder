```python
import requests # for making standard html requests
from bs4 import BeautifulSoup # magical tool for parsing html data
import json # for parsing data
from pandas import DataFrame as df # premier library for data organization
import re
```

```python
URL = 'https://arxiv.org/list/cs/new'
page = requests.get(URL)
page.encoding = 'ISO-885901'
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
info_list = soup.find_all('dl')
# print(info_list)
```

```python
articles_list = soup.find_all('dl')
```

```python
count = 0 

for cat_article_list in articles_list:
    for article in cat_article_list:
        print(article)
        count += 1
        print(count)
        if count > 11: 
            break
    break
```

```python
dl_data = soup.find_all('dl')

all_articles = {}
for dlitem in dl_data: 
    for dt_item, dd_item in zip(dlitem.find_all('dt'), dlitem.find_all('dd')):
        '''^\[.+\]$ matches any string with a "[" at the beginning, one or more characters, and a "]" at the end'''
        hash_num = int(json.loads((dt_item.find('a', text=re.compile("^\[.+\]$")).text))[0])
        title = (dd_item.find('div', class_='list-title mathjax').text.split('Title: ')[1]).strip()
        try:
            url = 'https://arxiv.org'.join(dt_item.find('a', title="Download PDF", href=True)['href'])
        except TypeError: 
            print(f"Title: '{title}' has no PDF download...")
        all_articles[hash_num] = {'title': title, 'url': url}
        
# print(all_articles)
```
